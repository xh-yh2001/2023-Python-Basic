```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-1, 1, 100)
y = 1.5*x*x + 0.2 + 0.2*np.random.randn(100)
plt.scatter(x, y)
plt.plot(x, 1.5*x*x + 0.2, 'r-', lw=3)
plt.show()
```

```python
# 快速梯度下降函数
def step_gradient_fast(b_current, w_current, points, learningRate):
    N = float(len(points))
    y = points[:,1]
    x = points[:,0]
    b_gradient = np.sum(-(2/N)*(y - ((w_current * x*x) + b_current)))
    w_gradient = np.sum(-(2/N)*x*x * (y - ((w_current * x*x) + b_current)))
    new_b = b_current - (learningRate * b_gradient)
    new_w = w_current - (learningRate * w_gradient)
    return [new_b, new_w]
```

```python
# 初始化参数
learning_rate = 0.1
b = 0
w = 0
num_iterations = 1001
points = np.zeros((100, 2))
points[:,0] = x
points[:,1] = y

# 迭代更新参数
for i in range(num_iterations):
    if i % 100 == 0:
        print('iteration %d, error is %.4f' % (i, np.mean((y - (w * x*x + b)) ** 2)))
        plt.scatter(x, y)
        plt.plot(x, w*x*x + b , 'r-', lw=2)
        plt.show()
    b, w = step_gradient_fast(b, w, points, learning_rate)
print("b = %.4f, w = %.4f" % (b, w))
```
